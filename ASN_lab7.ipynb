```python
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
# 1. Генеруємо сигнал
t = np.linspace(0, 1, num=1024)
signal = np.sin(32 * np.pi * t) + np.cos(64 * np.pi * t)
```

Нажаль базовий Wavelet клас має у собі баг, що не дозволяє використовувати його.  Розробники видалили з класу Wavelet функцію complex_cwt, але при цьому використовують її у стандартній функції pywt.cwt() , тому у якості прикладу було використано pywt.ContinuousWavelet та "mexh" вейвлет перетворення.

```python
# 2. Вибираємо базову вейвлет-функцію та рівні розшарування
wavelet = pywt.ContinuousWavelet('mexh')
levels = 5
```

```python
# 3. Побудова вейвлет-скейлограми
coef, freqs = pywt.cwt(signal, np.arange(1, levels+1), wavelet, sampling_period=1/1024, method='fft')
plt.imshow(coef, extent=[0, 1, freqs[-1], freqs[0]], cmap='PRGn', aspect='auto')
plt.colorbar()
plt.title('Wavelet Scalogram')
plt.xlabel('Time [sec]')
plt.ylabel('Frequency [Hz]')
plt.show()
```
